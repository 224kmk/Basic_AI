## python 특징

- **플랫폼 = os**
-- 윈도우, 리눅스,안드로이드, 맥OS, iOS 등 프로그램이 실행되는 운영 체제를 플랫폼이라고함.

- **독립적인 = 상관없는, 관계없는**
-- OS에 상관없이 한번 프로그램을 작성하면 사용가능

- **인터프리터 = 통역기를 사용하는 언어**
-- 소스코드를 바로 실행할 수 있게 지원하는 프로그램 실행 방법

- **객체 지향 동적 타이핑** 언어
-- **객체 지향적 언어** : 실행 순서가 아닌 단위 모듈(객체) 중심으로 프로그램을 작성 - 하나의 객체는 어떤 목적을 달성하기 위한 행동(method)와 속성(attribute)을 가지고있음. ex) 프린트 모듈, 입력 모듈, 처리 모듈을 합쳐서 하나로 만드는것.
-- **동적 타이핑** 언어 : 프로그램이 실행하는 시점에 프로그램이 사용해야할 데이터에 대한 타입을 결정함. ex) 실행하라고 입력했을때 실행하는 것.

### +  컴파일러 vs 인터프리터
- 작동방식
-- 컴파일러 : 소스코드를 기계어로 먼저 번역, 해당 플랫폼에 최적화 되어 프로그램을 실행
-- 인터프리터 : 별도의 번역과정 없이 소스코드를 실행시점에 해석하여 컴퓨터가 처리할 수 있도록 함

- 장점 / 단점
-- 컴파일러 : 실행속도가 빠름 / 한번에 많은 기억장소 필요
-- 인터프리터 : 간단히 작성, 메모리가 적게 필요 / 실행 속도가 느림

- 주요언어
-- 컴파일러 : C, 자바 ,C++, C#
-- 인터프리터 : 파이썬

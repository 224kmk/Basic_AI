## variables / memory
### 변수의 개요
- 가장 기초적인 프로그래밍 문법 개념
- 데이터(값)을 저장하기 위한 메모리 공간의 프로그래밍 상 이름
- 변수는 메모리 주소를 가지고 있고, 변수에 들어가는 값은 메모리 주소에 할당됨
-- ex) professor = "minki"의 뜻 ==> professor에 minki를 넣어라 ==> 정확히는 professor라는 변수에 "minki"라는 값을 넣으라는 의미임.
-- ex) A = 8 의 의미는 "A는 8이다"가 아닌 A라는 이름을 가진 메모리 주소에 8을 저장하라 임. 파이썬 -> OS -> Memory

### 변수 이름 작명법
- 알파벳, 숫자, 언더스코어(_)로 선언가능
-- ex) data=0, _a12=2, _gg='afdf'
- 변수명은 의미 있는 단어로 표기하는 것이 좋음
-- ex) student_name="minki"
- 변수명은 대소문자가 구분됨.
-- ex)ABC와 Abc는 다름
- 특별한 의미가 있는 예약어는 쓰지않음
-- ex) for, if, else 등

#### 2. 연산자와 피연산자
- +, -, *, / 같은 기호들을 연산자
- 연산자에 의해 계산이 되는 숫자들은 피연산자
- "3+ 2" 에서 3,2는 피연산자, +는 연산자
- 문자간에도 + 연산이 가능 -> concatenate "a" + "b" = ab
- ** = 제곱승 % = 나머지

```python
a=1 # 변수 a에 1을 할당
a = a+1 # a에 1을 더한 후 그 값을 a에 할당
print(a) # a출력

# a 증가 연산
a += 1 # 좌변의 a는 저장되는 공간 우변의 a는 메모리에서 가져오는 값.

# a 감소 연산
a -= 1
```

```python
# 데이터 형 변횐 : 정수형 <-> 실수형
# 정수형 -> 실수형
a=10
print(type(a))
print(float(a)) # 찍히는건 10.0이지만 아직 a는 integer임
print(type(a))
a = float(a)
type(a)

# 실수형 -> 정수형
b=10.7
c=10.3
b=int(b) # 형변환 뒤의 소수점이 사라짐
c=int(c) # 마찬가지
print(b+c) # 그래서 20 출력

# 문자형 -> 실수형
d = "76.3"
print(type(d))
d = float(d)
type(d)

e = "15.5"
e + e # concatenate가 됨
```

### List, Array
- 시퀀스 자료형, 여러 데이터들의 집합
- int, float 같은 다양한 데이터 타입 포함

#### 특징
- 인덱싱
- 슬라이싱
- 리스트 연산
- 추가 삭제
- 메모리 저장 방식
- 패킹과 언패킹
- 이차원 리스트


```python
# 인덱싱(indexing) : list에 있는 값은 주소(offset)을 가짐
colors = ["red", "blue", "green"]
len(colors)
```

```python
# 슬라이싱(slicing) : list의 값을 잘라서 쓰는것. list주소 값을 기반으로 부분 값을 반환
# 처음은 0
colors = ["red", "blue", "green", "black", "gray", "white", "orange", "pink"]
print(colors[5:11])
print(colors[-9:]) # 맨 오른쪽이 -1임
print(colors[0:10:2]) # 2칸씩 짤라서 [시작:끝:step]
print(colors[::-1]) # 거꾸로 한칸씩
```

```python
# 리스트 연산
color1 = ["red", "blue", "green", "black"]
color2 = ["gray", "white", "orange", "pink"]
color1 + color2 # concatenate
print(color1[0])
print("blue" in color1) # color1에 블루가 들어있나라고 물은것
print(color1 * 2)
total_color = color1 + color2
print(total_color)
```

```python
# 추가 삭제
color1.extend(["purple"]) # 추가 [] 주의
print(color1)
color1.remove("red") # 제거 []없음 안의 요소를 제거하는것
print(color1)
```

```python
a = [5, 4, 3, 2, 1]
b = [1, 2, 3, 4, 5]
b = a # b에 a를 넣음 즉 a로 바뀜 같은 메모리 공간을 가짐.
print(b)
print("\n")

a.sort() # 오름차순 정렬을 했을때 b와 a가 같은 주소를 가르키기때문에 b도 정렬이 됨
print(a)
print(b) # a만 sort했는데 변하는것을 볼 수 있음
print("\n")

a[2] = 100 # 마찬가지로 a의 3번째 변수를 변경하여도 b의 변수도 변경
print(a)
print(b)
print("\n")

b = [1,2,3,4,5] # 다시 선언하여 메모리 주소를 변경
print(a)
print(b)
print("\n")

# 복사를 하면서 메모리 공간을 분리하게 되려면 슬라이싱을 사용
b = a[:]
print(a)
print(b)
a[3] = 50
print(a)
print(b) # 바뀌지 않은 모습을 볼 수 있음
```

```python
# 패킹과 언패킹
# 패킹 : 한 변수에 여러 개의 데이터를 넣는것
# 언패킹 : 한 변수의 데이터를 각각의 변수로 반환
t = [1, 2, 3] # 1,2,3 변수를 t에 패킹
a, b, c = t # t에 있는 값 1, 2, 3을 변수 a, b, c 에 언패킹
print(t, a, b, c) # 리스트와 요소가 출력
```

```python
kor_score=[10,20,30]
math_score=[40,50,60]
eng_score=[70,80,90]
midterm_score =[kor_score, math_score, eng_score]
midterm_score
```

```python
# 기존 1D 리스트 복사하는 것 처럼 복사하면 2D 리스트에서는 같이 바뀜 주소값이 같이 오기때문
# 기존방법말고 다른방법으로 복사해야함
midterm_copy = midterm_score[:]
print(midterm_copy) # 복사는 됨
print("\n")
midterm_score[0][2] = 75 # 기존 점수 변경
print(midterm_score)
print(midterm_copy) # 같이 변경되는것을 볼 수 있음
```

```python
# 2D이상에서 카피하는 방법
import copy
midterm_copy2 = copy.deepcopy(midterm_score) # 복사
midterm_score[0][1] = 55 # 값변경
print(midterm_score)
print(midterm_copy2) # 변경되지 않은것을 볼 수 있음
```

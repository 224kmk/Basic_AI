## 함수
- 어떤 일을 수행하는 코드의 덩어리
- 반복적인 수행을 1회만 작성 후 호출

- 캡슐화 : 인터페이스만 알면 타인의 코드 사용

### 함수 선언
- 함수이름, parameter, indentation, return value(optional)
-- ex) :
        def 함수 이름 (parameter #1, ...,):
            수행문 #1(statements)
            수행문 #2(statements) ...
            return <반환값>

```python
def calculate_rectangle_area(x , y): #3
    return x * y
rectangle_x = 10
rectangle_y = 20
print ("사각형 x의 길이: ", rectangle_x) # 1 #4
print ("사각형 y의 길이: ", rectangle_y)
# 넓이를 구하는 함수 호출
print ("사각형의 넓이: ", calculate_rectangle_area(rectangle_x, rectangle_y)) #2

# 메모리에 코드가 올라감
# 메인 프로그램 수행(#1) -> 함수 호출(#2) -> 함수 수행(#3) ->  메인 프로그램 수행(#4)
```

```python
def f(x):            # x는 parameter
    return x*2 + 7   # 2x + 7 return 값

def g(x):
    return x**2

x = 2               # argument : 실제 parameter에 대입된 값
result = f(x) + g(x) + f(g(x)) +g(f(x))
result
```

### parameter와 return의 유무에 따라 함수를 4가지를 만들 수 있다.
- parameter가 존재, 반환값 존재 : parameter를 사용하여 수행문 수행 후 결과값 반환 => 일반적인 함수
- parameter가 존재, 반환값 없음 : parameter를 사용 수행문만 수행 
- parameter가 없음, 반환값 없음 : 함수 내의 수행문만 수행
- parameter가 없음, 반환값 존재 : parameter없이, 수행문 수행 후 결과값 반환


```python
def f(x):
    print(x +10)
f(10)
# 20이 찍힘 

c = f(10)
# c의 값은 20이라고 찍히지만

print(c) # c자체의 값은 없다. return 값이 없기때문
```

```python
def g(x):
    return x + 10
c = g(10) # 화면에 찍힐뿐임
print(c) # 할당된 것
g(10)  # 화면에 찍힐뿐임
```

```python
# print formatting
# %s 문자열 (String)
# %c 문자 1개(character)
# %d 정수 (Integer)
# %f 부동소수 (floating-point)
# %o 8진수
# %x 16진수
# %% Literal % (문자 % 자체)
print(1,2,3)
print("a" + " " + "b" + " " + "c")
print("%d %d %d" % (1,2,3))
print("{} {} {}".format("a","b","c"))
print("\n")

# old-school formatting : 일반적으로 % -format, str.format을 사용
print('%s %s' % ('one', 'two')) 
print('{} {}'.format('one', 'two')) 
print('%d %d' % (1, 2)) # d
print('{} {}'.format(1, 2))
print("\n")

# %-format : %datatype, %(variable)
print("I eat %d apples." % 3)
print("I eat %s apples." % "five")
number = 3; day="three"
print("I ate %d apples. I was sick for %s days." % (number, day))
print("Product: %s, Price per unit: %f." % ("Apple", 5.243)) 
print("Product: %10s, Price per unit: %10.1f." % ("Apple", 5.243)) # 10칸 확보 후 소수점 한자리 까지 출력
print("\n")

# “~~~~{datatype}~~~~”.format(argument) 
age = 36; name='Sungchul Choi'
print("I’m {0} years old.".format(age))
print("My name is {0} and {1} years old.".format(name,age)) # format을 사용하는 방법
print("Product: {0}, Price per unit: {1:.3f}.".format("Apple", 5.243))
```

```python
# padding
# 여유 공간을 지정해서 글자배열과 소수점 자리 맞추기
print("Product: %5s, Price per unit: %.5f." % ("Apple", 5.243))
print("Product: {0:5s}, Price per unit: {1:.5f}.".format("Apple", 5.243))
print("Product: %10s, Price per unit: %10.3f." % ("Apple", 5.243))
print("Product: {0:<10s}, Price per unit: {1:10.3f}.".format("Apple", 5.243)) # >와 < 를 이용해서 좌우로 붙일수있음.
```

```python
# naming
# 해당 표시할 내용을 변수로 표시하여 입력
print("Product: %(name)10s, Price per unit: %(price)10.5f." % {"name":"Apple", "price":5.243}) # name을 변수로 입력하여 넣음
print("Product: {name:>10s}, Price per unit: {price:10.5f}.".format(name="Apple", price=5.243))
```

```python
# f-string
# python 3.6 이후, PEP498에 근거한 formatting 기법
# 기본이 왼쪽 정렬임
name = "Sungchul"
age = 39
print(f"Hello, {name}. You are {age}.")
print(f'{name:20}')
print(f'{name:>20}')
print(f'{name:*<20}') # 왼쪽 정렬 후 나머지는 *로
print(f'{name:*>20}') # 오른쪽 정렬 후 나머지는 *로
print(f'{name:*^20}') # 가운데 정렬 후 나머지는 *로
number = 3.141592653589793
print(f'{number:.2f}') # 넘버의 소수점 둘째자리까지.
```

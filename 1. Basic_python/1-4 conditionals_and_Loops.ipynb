## 1-4 conditionals & Loops
- 조건에 따라 특정한 동작을 하게하는 명렁어
- if, else, elif등 예약어를 사용

- 1. 조건 판단 방법
- 2. 조건 일치 시 수행 명령 block ":"와 들여쓰기
- 3. 조건 불일치 시 수행 명령 block


```python
# if <조건>: # if를 쓰고 조건 삽입 후 “:” 입력
   # <수행 명령1-1> # 들여쓰기(indentation)후 수행명령 입력
   # <수행 명령1-2> # 같은 조건하에 실행일 경우 들여쓰기 유지
   
# else: # 조건이 불일치할 경우 수행할 명령 block
   # <수행 명령2-1> # 조건 불일치 시 수행할 명령 입력
   # <수행 명령2-2> # 조건 불일치 시 수행할 명령 들여쓰기 유지
```

```python
print ("Tell me your age?")
myage = int(input()) # 나이를 입력 받아 myage 변수에 할당
if myage < 30: # myage 가 30 미만일 때
    print ("Welcome to the club")
else: # myage 가 30 이상일 때
    print ("Oh! No. You are not accepted.")
```

```python
# 참/거짓을 구분하기 위해 비교 연산자 사용

# x < y ~보다 작음 x과 y보다 작은지 검사
# x > y ~ 보다 큼 x과 y보다 큰지 검사
# x == y 같음 x와 y과 같은지 검사
# x != y 같지 않음 x와 y과 다른지 검사
# x is y (값과 메모리 주소)     ==> 메모리의 주소를 비교하기 때문에 값이 같더라도 다를 수 있음
# x is not y (값과 메모리 주소) ==> 그러므로 주의해야함
# x >= y 크거나 같음 x과 y보다 이상인지 검사
# x <= y 작거나 같음 x과 y보다 이하인지 검사
```

```python
# 숫자형의 경우는 수학에서의 참/거짓과 동일
# 컴퓨터는 존재하면 참 없으면 거짓이라고 판단함
# 존재하기때문에 hello가 출력
if 1:
    print("hello")
else:
    print("goodbye")
```

```python
# 논리 키워드 : and, or, not
a = 8
b = 5
print(a == 8 and b == 4) # 거짓
print(a > 7 or b > 7) # 참
not (a > 7) # 거짓, a>7 인 것이 참 이므로 거짓으로 판단됨
```

```python
# 삼항연산자
value = 12
is_even = True if value % 2 == 0 else False
print(is_even)
# if의 말을 만족하면 True 아니면 False
```

- 0부터 시작
- 무한 loop 주의


```python
for i in range(0,5):
    print("hi")
```

```python
# 요소 출력
for i in "abcdefg":
    print (i)
```

```python
# 리스트 출력
for i in ["americano", "latte", "frafuchino"]:
      print (i)
```

```python
# 간격 두고 세기
for i in range(1, 10, 2):
    # 1부터 10까지 2씩 증가시키면서 반복문 수행
    print (i)
print("\n")
# 역순
for i in range(10, 1, -1):
    # 10부터 1까지 -1씩 감소시키면서 반복문 수행
    print (i)
```

```python
# for 와 while은 반복횟수를 정확히 알고 모를때 사용한다
# for를 사용할땐 횟수를 알때
# while을 사용할땐 횟수를 모를때

i = 1           # ① i 변수에 1 할당
while i < 10:   # ② i가 10 미만인지 판단
    print (i)   # ③ 조건에 만족할 때 i 출력, i에 1을 더함
    i += 1      # ④ i가 10이 되면 반복 종료
```

```python
# 반복의 제어 - break, continue
# break - 만나면 조건 종료
for i in range(10):
    if i == 5: break # i가 5가 되면 반복 종료
    print (i)
print("EOP") # 반복 종료 후 “EOP” 출력
print("\n")
# continue - 만나면 다시 올라감
for i in range(10):
    if i == 5: continue # i가 5가 되면 i를 출력하지 않음
    print (i)
print("EOP") # 반복 종료 후 “EOP” 출력
```

```python
# 10진수 2진수 변환
print ("input decimal number: ",)
decimal = int(input())
result =""
loop_counter = 0
while (decimal > 0):
    temp_decimal_input = decimal
    temp_result_input=result

    remainder = decimal % 2
    decimal = decimal // 2
    result = str(remainder) + result

    print ("-----------", loop_counter, "loop value check -------- ")
    print ("Initial decimal:", temp_decimal_input,
           ", Remainder:",remainder,
           ", Initial result", temp_result_input)
    print ("Output decimal:", decimal,
            "Output result:", result)
    print ("------------------------------------------------------")
    print ("")
    
    loop_counter += 1
print ("Binary number is", result)
```

```python
# 이차원 리스트처리
# 사람별 평균 구하기
kor_score = [49,79,20,100,80]
math_score = [43,59,85,30, 90]
eng_score = [49,79,48,60,100]
midterm_score = [kor_score, math_score, eng_score]
print (midterm_score[0][2])

student_score = [0,0,0,0,0]
i = 0
for subject in midterm_score:
    for score in subject:
        student_score[i] += score # 각 학생마다 개별로 교과 점수를 저장
        i   += 1 # 학생 index 구분
    i = 0 # 과목이 바뀔 때 학생 인덱스 초기화
else:
    a, b, c, d, e = student_score # 학생 별 점수를 unpacking
    student_average = [a/3,b/3,c/3,d/3,e/3]
    print (student_average)
```

### 함수 호출 방식 개요
- parameter를 전달하는 방식
-- 1. 값에 의한 호출(Call by Value)
-- 2. 참조의 의한 호출(Call by Reference)
-- 3. 객체 참조에 의한 호출(Call by Object Reference)

- 1. Call by Value
-- 함수에 인자를 넘길 때 값만 넘김
-- 함수 내에 인자 값 변경 시, 호출자에게 영향을 주지 않음

- 2. Call by Reference
-- 함수에 인자를 넘걸 때 **메모리 주소**를 넘김
-- 함수 내에 인자 값 변경 시, **호출자**의 값도 변경

- 3. 파이썬 함수 호출 방식 - Call by Object Reference
-- 파이썬은 객체의 주소가 함수로 전달되는 방식
-- 전달된 객체를 참조하여 변경 시 호출자에게 영향을 주나, 새로운 객체를 만들 경우 호출자에게 영향을 주지 않음.


```python
def spam(eggs):
    eggs.append(1) # 기존 객체의 주소값에 [1] 추가
    eggs = [2, 3] # 새로운 객체 생성
    print(eggs)

ham = [0]
spam(ham)
print(ham) # [0, 1]
# 햄의 객체가 살아서 eggs로 넘어가고 append로 인해 ham에 추가되어 0,1 이 된다.
# 하지만 eggs를 다시 정의해서 기존의 주소가 끊겨서 eggs는 2,3으로 변경된다.
```

```python
# swap
# 함수를 통해 변수 간의 값을 교환하는 함수

# value만 교환
def swap_value (x, y):
    temp = x
    x = y
    y = temp

# 주소만 변환
def swap_offset (offset_x, offset_y):
    temp = a[offset_x]
    a[offset_x] = a[offset_y]
    a[offset_y] = temp

# 리스트를 변환
def swap_reference (list, offset_x, offset_y):
    temp = list[offset_x]
    list[offset_x] = list[offset_y]
    list[offset_y] = temp
```

```python
a = [1,2,3,4,5]

# 값 변환 
swap_value(a[1], a[2])  # 1의 값과 2의 값을 바꿔줘
print(a) # [1,2,3,4,5] # 하지만 값은 그대로 나옴 ==> 주소의 값이 고정되어 있기때문에 바뀌지않음

# 주소 변환
swap_offset(1,2)        # 1과 2의 주소값을 변경해줘
print(a) # [1,3,2,4,5] # 위치의 주소값을 변경했기때문에 값도 변경되는 모습을 볼 수 있음

# 리스트 변환
swap_reference(a, 1, 2) # 리스트에서 1과 2의 값을 변경해줘
print(a) # [1,3,2,4,5]  # 리스트에서 값 자체가 변경되기때문에 변경됨.
```

```python
# 재귀함수
# 자기 자신을 도는 함수
# 점화식
def factorial(n):
    if n == 1:
        return 1
    else:
        return n + factorial(n-1)
print (factorial(int(input("Input Number for Factorial Calculation: "))))
```

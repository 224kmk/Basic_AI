```python
# split
# string type의 값을 “기준값”으로 나눠서 List 형태로 변환
items = 'zero one two three'.split() # 빈칸을 기준으로 문자열 나누기
print(items)

example = 'python,java,javascript' # ","을 기준으로 문자열 나누기
print(example.split(","))

a, b, c = example.split(",") # 리스트에 있는 각 값을 a,b,c 변수로 unpacking
example = 'teamlab.technology.io'
subdomain, domain, tld = example.split('.') # "."을 기준으로 문자열 나누기 → Unpacking
```

```python
# join
# String으로 구성된 list를 합쳐 하나의 string으로 반환
colors = ['red', 'blue', 'green', 'yellow']
result = ''.join(colors)
print(result)

result = ' '.join(colors) # 연결 시 빈칸 1칸으로 연결
print(result)

result = ', '.join(colors) # 연결 시 ", "으로 연결
print(result)

result = '-'.join(colors) # 연결 시 "-"으로 연결
print(result)
```

```python
# list comprehension
# List 사용하여 간단히 다른 List를 만드는 기법
# 포괄적인 List, 포함되는 리스트라는 의미로 사용됨
# 파이썬에서 가장 많이 사용되는 기법 중 하나
# 일반적으로 for + append 보다 속도가 빠름
# ex) 1
result = []
for i in range(10):
    result.append(i)

result
```

```python
result = [i for i in range(10)]
print(result)

result = [i for i in range(10) if i % 2 == 0]
result
```

```python
# ex) 2
word_1 = "Hello"
word_2 = "World"
result = [i+j for i in word_1 for j in word_2] # Nested For loop 총 25개가 나옴
result
```

```python
# ex)3
case_1 = ["A","B","C"]
case_2 = ["D","E","A"]
result = [i+j for i in case_1 for j in case_2]
print(result)

result = [i+j for i in case_1 for j in case_2 if not(i==j)]
# Filter: i랑 j과 같다면 List에 추가하지 않음
# [i+j if not(i==j) else i for i in case_1 for j in case_2]
print(result)

result.sort()
print(result)
```

```python
# ex) 4
words = 'The quick brown fox jumps over the lazy dog'.split()
# 문장을 빈칸 기준으로 나눠 list로 변환
print (words)
print("\n")
stuff = [[w.upper(), w.lower(), len(w)] for w in words]
# list의 각 elemente들을 대문자, 소문자, 길이로 변환하여 two dimensional list로 변환
for i in stuff:
    print(i)
```

```python
# Two dimensional vs One dimensional
case_1 = ["A","B","C"]
case_2 = ["D","E","A"]

result = [i+j for i in case_1 for j in case_2] # 1차원으로 생성
print(result)

result = [ [i+j for i in case_1] for j in case_2] # 2차원으로 생성 중간에 []가 하나 있음
print(result)
```

```python
# enumerate & zip
# enumerate : list의 element를 추출할 때 번호를 붙여서 추출
my_str = "ABCD"
print({v : i for i, v in enumerate(my_str)}) # 

for i, v in enumerate(['tic', 'tac', 'toe']): # list의 있는 index와 값을 unpacking
    print(i, v)
print("\n")

mylist = ['a', 'b', 'c', 'd']
print(list(enumerate(mylist))) # list의 있는 index와 값을 unpacking하여 list로 저장
print("\n")

set_text = 'Artificial intelligence (AI), is intelligence demonstrated by machines, unlike the natural intelligence displayed by humans and animals.'.split()
{i:j for i,j in enumerate(set_text)}
# 문장을 list로 만들고 list의 index와 값을 unpacking하여 dict로 저장
```

```python
# zip
# zip : 두 개의 list의 값을 병렬적으로 추출함
alist = ['a1', 'a2', 'a3'] 
blist = ['b1', 'b2', 'b3']
for a, b in zip(alist, blist): # 병렬적으로 값을 추출
    print (a,b)

print("\n")
a,b,c =zip((1,2,3),(10,20,30),(100,200,300)) #각 tuple의 같은 index 끼리 묶음
print([sum(x) for x in zip((1,2,3), (10,20,30), (100,200,300))])
# 각 Tuple 같은 index를 묶어 합을 list로 변환
```

```python
 # enumerate & zip 동시 사용 예
alist = ['a1', 'a2', 'a3']
blist = ['b1', 'b2', 'b3']
for i, (a, b) in enumerate(zip(alist, blist)):
    print(i, a, b) # index alist[index] blist[index] 표시
```

```python
# lambda & map & reduce
# lambda
# 함수 이름 없이, 함수처럼 쓸 수 있는 익명함수
# 수학의 람다 대수에서 유래함
def f(x, y):
    return x + y
print(f(1, 4))
# 위와 아래 동일한 함수임
f = lambda x, y: x + y
print(f(1, 4))
```

```python
# map function
# 두 개 이상의 list에도 적용 가능함, if filter도 사용가능
# 실행시점의 값을 생성, 메모리 효율적
ex = [1,2,3,4,5]
f = lambda x, y: x + y
print(list(map(f, ex, ex)))

# list를 붙여야함
list(map(lambda x: x ** 2 if x % 2 == 0 else x,ex))
```

```python
# reduce
# map function과 달리 list에 똑같은 함수를 적용해서 통합
from functools import reduce
print(reduce(lambda x, y: x+y, [1, 2, 3, 4, 5]))
# 통합하는과정이란 1+2+3+4+5가 순차적으로 더해지는것임.

# 단점 : 코드의 직관성이 떨어짐
# Legacy library나 머신러닝 코드에서는 여전히 사용중
```

```python
# iterable object
# Sequence형 자료형에서 데이터를 순서대로 추출하는 object
# 내부적 구현으로 __iter__ 와 __next__ 가 사용됨
# iter() 와 next() 함수로 iterable 객체를 iterator object로 사용

cities = ["Seoul", "Busan","Jeju"]
iter_obj = iter(cities)

print(next(iter_obj))
print(next(iter_obj))
print(next(iter_obj))
next(iter_obj) # 없어서 출력되지 않음

# 메모리 주소를 저장하는데 메모리 주소가 연속적으로 있는것을 말함
# cities -> seoul, busan, jeju가 순차적으로 되어있어서 전체를 기억하지는 못하고 다음것만 기억함
```

```python
# generator
#iterable object를 특수한 형태로 사용해주는 함수
# element가 사용되는 시점에 값을 메모리에 반환
# yield를 사용해 한번에 하나의 element만 반환함

# 일반적인 함수
def general_list(value):
    result = []
    for i in range(value):
        result.append(i)
    return result
print(general_list(50))
```

```python
# 크기 구하기
import sys
result = general_list(50)
sys.getsizeof(result)
```

```python
# 제너레이터 함수
def generator_list(value):
    result = []
    for i in range(value):
        yield i
```

```python
# 사용되는 시점에 메모리를 반환하기때문에 크기가 작은것을 볼수 있음
result = generator_list(50)
sys.getsizeof(result)
```

```python
# list comprehension과 유사한 형태로 generator형태의 list 생성
# generator expression 이라는 이름으로도 부름
# [ ] 대신 ( ) 를 사용하여 표현
# 메모리를 아껴야되는 부분에서 사용할 수 있음
gen_ex = (n*n for n in range(500))
print(type(gen_ex)) 
```

```python
# list 타입의 데이터를 반환해주는 함수는 generator로 만들자
# 읽기 쉬운 장점, 중간 과정에서 loop 이 중단될 수 있을 때!
# 큰 데이터를 처리할 때는 generator expression을 고려하자
# 데이터가 커도 처리의 어려움이 없음
# 파일 데이터를 처리할 때도 generator를 쓰자
```

### fuction passing arguments
- 함수에 입력되는 arguments의 다양한 형태
- 1) Keyword arguments
- 2) Default arguments
- 3) Variable-length arguments

```python
# 1. Keyword arguments
# 함수에 입력되는 parameter의 변수명을 사용, arguments를 넘김
def print_somthing(my_name, your_name, third_name):
    print("Hello {0}, My name is {1}".format(your_name, my_name))

print_somthing("abc", "Sungchul", "TEAMLAB") # 첫번째에 있는 argument가 abc이기때문에 abc가 출력이 됨
print_somthing(third_name="abc", your_name="TEAMLAB", my_name="Sungchul")
```

```python
# 2. Default arguments
# parameter의 기본 값을 사용, 입력하지 않을 경우 기본값 출력
def print_somthing_2(my_name, your_name="TEAMLAB"): # your_name=TEAMLAB이 사라지면 입력한 값이 출력
    print("Hello {0}, My name is {1}".format(your_name, my_name))

print_somthing_2("Sungchul", "TEAMLAB")
print_somthing_2("Sungchul") # 지정한값을 따로 넣지않아도 기본값 출력
```

```python
# 가변인자 (variable-length)
# 개수가 정해지지 않은 변수를 함수의 parameter로 사용하는 법
# Keyword arguments와 함께, argument 추가가 가능
# Asterisk(*) 기호를 사용하여 함수의 parameter를 표시함
# 입력된 값은 tuple type으로 사용할 수 있음
# 가변인자는 오직 한 개만 맨 마지막 parameter 위치에 사용가능

# 가변인자는 일반적으로 *args를 변수명으로 사용
# 기존 parameter 이후에 나오는 값을 tuple로 저장함
def asterisk_test(a, b, *args):
    print(list(args))
    print(type(args))
    return a+b+sum(args)
print(asterisk_test(1, 2, 3, 4, 5))
```

```python
# unpacking도 가능함
def asterisk_test_2(*args):
    x, y, z = args
    return x, y, z
print(asterisk_test_2(3, 4, 5))
```

```python
# 키워드 가변인자 (Keyword variable-length )
# variable-length asterisk
# Parameter 이름을 따로 지정하지 않고 입력하는 방법
# asterisk(*) 두개를 사용하여 함수의 parameter를 표시함
# 입력된 값은 dict type으로 사용할 수 있음
# 가변인자는 오직 한 개만 기존 가변인자 다음에 사용

def kwargs_test_1(**kwargs):
    print(kwargs)
    print(type(kwargs))

def kwargs_test_2(one, two=3, *args, **kwargs):
    print(one+two+sum(args))
    print("one :", one)
    print("two :", two)
    print("args :", args)
    print("kwargs :", kwargs)

print(kwargs_test_1(first=3, second=4, third=5, d= 10))
print("\n")
kwargs_test_2(10,30, 3,5,6,7,8, first=3, second=4, third=5)
# 10은 one에 / 30은 two에 디폴트 값인 3이있지만 무시됨 / args 가변인자에 3,5,6,7,8 / 키워드 가변인자에는 3,4,5가 들어감
# 순서를 막 적게 되면 오류 발생
```

```python
# asterisk
# 흔히 알고 있는 * 를 의미함
# 단순 곱셈, 제곱연산, 가변 인자 활용 등 다양하게 사용됨
```

```python
# asterisk – unpacking a container
def asterisk_test(a, *args): # 여러개의 가변인자를 받을수 있는 astrick
    print(a, args)
    print(a, *args)
    print(type(args))

test = (2,3,4,5,6)
asterisk_test(1, *test) # unpacking의 astrick
# 1값은 a로 들어감 
# test의 값은 args로 들어감 함수안에서 *를 넣게 되면 unpacking이 일어남
# 그래서 결과값에 그냥 args는 튜플로 묶여있는것을 볼 수 있고
# 별표가 들어가게되면 튜플형태가 풀리는 것을 볼 수 있다.

print("\n")
print(asterisk_test(1, test))
# 그냥 *를 안붙이면 unpacking이 없어서 그냥 됨.
```

```python
## 키워드 언패킹
def asterisk_test(a, b, c, d,):
    print(a, b, c, d)

data = {"b":1 , "c":2, "d":3}
asterisk_test(10, **data) # astrick가 두개인 모습을 볼 수 있음.
# 처음에 10
# 두번째에는 1,2,3이 들어가야하지만 키워드로 묶여있던것이 풀림.
```
